```python
import pandas as pd
df = pd.read_csv("C:/Users/eliza/Documents/Semester V/Sentiment Analysis/CITY/neg_df.csv")
df
```

```python
# Convert to lowercase for consistent matching
df['business_name'] = df['business_name'].str.lower()

# Define regex patterns
patterns = [
    r'dollar\s+general',
    r"tobacco\s+joes.",
    r'safe\s+family\s+defense',
    r'family\s+video',
    r'walgreens',
    r'sweetgreen',
    r'proteinhouse',
    r'river\s+city\s+environment',
    r"trader\s+joe's",
    r'\bgrocery\b',                         # word "grocery" anywhere
    r'\bmarket\b',                          # word "market" anywhere
    r'\bpet\b',
    r'\bvet\b',      # word "pet" anywhere
    r'\bveterinarian\b',
    r'\bspa\b',
    r'\bveterinary\b',  
    r'\bjuicery\b',  
    r'\bgym\b', 
    r'\bfitness\b',
    r'\banimal\b',
    r'\bfood\b',
    r'\bclub\b',
    r'\bkangen\b',
    r'\bboutique\b',
    r'\bmassage\b',
    r'\bair\b',
    r'\bcampus\b',
    r'\bpaws\b',
    r'\bdog\b',
    r'\bcat\b',
    r'\bcampus\b',
    r'\bcamp\b',
    r'\bhair\b',
    r'\boutlet\b',
    r'\bnail\b',
    r'\bmuscle\b',
    r'\bservice\b',
    r'\bcampus\b',
    r'\bhaircut\b',
    r'\blawn\b',
    r'\bsports\b',
    r'\bwax\b'
    r'\bsalon\b'
    r'\brestaurant\b'
    r'\bpool\b'
]

# Combine patterns into one regex using |
combined_pattern = '|'.join(patterns)

# Filter rows where business_name matches the pattern
filtered_df = df[df['business_name'].str.contains(combined_pattern, na=False, regex=True)]

filtered_df
```

```python
df_cleaned = df.drop(filtered_df.index)
df_cleaned.reset_index(drop=True, inplace=True)
df_cleaned
```

```python
df_cleaned.to_csv("negative_df.csv", index = False)
```

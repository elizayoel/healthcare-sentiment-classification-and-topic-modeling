```python
from tensorflow import keras
import keras_tuner as kt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam

def build_model(hp):
    model = Sequential()

    # Embedding layer (fixed GloVe)
    model.add(Embedding(
        input_dim=VOCAB_SIZE,
        output_dim=100,
        weights=[embedding_matrix],
        input_length=MAX_LEN,
        trainable=True
    ))

    # Flatten the embedding layer output
    model.add(Flatten())

    # Dense Block 1
    model.add(Dense(
        hp.Choice('dense_1', [128, 256, 512]),
        activation='relu'
    ))
    model.add(Dropout(hp.Float('dropout_1', 0.3, 0.6, step=0.1)))

    # Dense Block 2
    model.add(Dense(
        hp.Choice('dense_2', [64, 128, 256]),
        activation='relu'
    ))
    model.add(Dropout(hp.Float('dropout_2', 0.3, 0.6, step=0.1)))

    # Dense Block 3
    model.add(Dense(
        hp.Choice('dense_3', [32, 64, 128]),
        activation='relu'
    ))

    # Output Layer
    model.add(Dense(1, activation='sigmoid'))

    # Compile
    model.compile(
        optimizer=Adam(hp.Choice('learning_rate', [1e-2, 1e-3, 1e-4])),
        loss='binary_crossentropy',
        metrics=METRICS
    )

    return model
```

**EARLY STOPPING**

this is a callback that monitors a metrix during training and stops the training early if it stops improving  
the best weight is automatically saved thorugh 'restore_best_weights', so no need to use the old callback function

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=2,
    restore_best_weights=True
)
```

**TUNER SETUP**

similar to sklearn randomsearch's setting, where we set n estimators and cv  
this one is for deep learning

1. max trials = like n_iter = number of random configs to try
2. execution per trial = like cv = how many times to train per config

keras tuner randomly samples 10 combinations of hyperparameters from the build_bilstm_model function  

for each combination:
- it builds the model with those specific settings
- it trains the model using fit() with training data
- it evaluates on validation set 

```python
tuner = kt.RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials= 10,
    executions_per_trial=1,
    directory='tuner_logs', #keras tuner will create this directory to store logs automatically
    project_name='DNN_tuning' #MUST CHANGE!!!
)
```

**RUN SEARCH**

1. keras tuner tries many combination of hyperparameters
2. for each combination, it builds a new model using that config
- trains the model for a limited number of epoch
- it saves the validation score but does not save the model permanently
3. after trying all combinations, it selects the best hp combination

NOTES
1. the model from tuning is not fully trained
2. it might have learned something, but its just a trial run
3. it might even be partially underfitted or overfitted depending on the randomseed, shuffle, early stopping, etc

```python
# Run search
tuner.search(
    X_train_pad, y_train,
    validation_data=(X_val_pad, y_val),
    epochs=5,
    batch_size=32,
    callbacks=[early_stopping],
    verbose=2
)
```

**BEST MODEL SELECTION AND FINAL TRAINING**

1. the partially trained model is not used
2. use the best hp found from tuning = best_hp
3. build a new fresh model with 
- fresh architecture
- brand new untrained weight, randomly initialized  or from pretrained embedding


NOTES:
1. hp = set by us or kerastuner before training begins. examples:
- number of lstm units
- dropout rate
- learning rate
- embedding size

2. weights = parameter  
learned by the model during training  
this are actual numerical values that are updated by backpropagation to minimize the loss function  
every neuron and connection has weights  

example = in a dense layer: output = input x weight + bias  

weights are randomly initialized at the start, unless u load pretrained weights  

weights = KNOWLEDGE A MODEL GAINS WHILE TRAINING  
the model knows nothing until the weights are trained

```python
best_hp = tuner.get_best_hyperparameters(1)[0]
best_model = tuner.hypermodel.build(best_hp)

history = best_model.fit(
    X_train_pad, y_train,
    validation_data=(X_val_pad, y_val),
    epochs=5,
    batch_size=32,
    callbacks=[early_stopping],
    verbose=2
)
```

```python
import matplotlib.pyplot as plt

def plot_graphs(history, metric):
    plt.plot(history.history[metric])
    plt.plot(history.history['val_' + metric])
    plt.xlabel("Epochs")
    plt.ylabel(metric)
    plt.legend([metric, 'val_' + metric])

#automate plotting of four subplots (acc, loss, prec, recall)
#use history object returned by model.fit()
#compares TRAINING and VALIDATION performance for each metric across epochs
def model_plot_graph(history):
    plt.figure(figsize=(16, 8))

    plt.subplot(2, 2, 1)
    plot_graphs(history, 'accuracy')
    plt.ylim(None, 1)

    plt.subplot(2, 2, 2)
    plot_graphs(history, 'loss')
    plt.ylim(0, None)

    plt.subplot(2, 2, 3)
    plot_graphs(history, 'precision')
    plt.ylim(None, 1)

    plt.subplot(2, 2, 4)
    plot_graphs(history, 'recall')
    plt.ylim(0, None)
    

model_plot_graph(history)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

def evaluate_model_sklearn(model, X, y, label="Test Set"):
    # Predict probabilities and round to binary (0 or 1)
    y_pred_prob = model.predict(X)
    y_pred = (y_pred_prob > 0.5).astype("int32")

    # Metrics
    acc = accuracy_score(y, y_pred)
    prec = precision_score(y, y_pred)
    rec = recall_score(y, y_pred)
    f1 = f1_score(y, y_pred)

    print(f" {label} Evaluation")
    print(f"  Accuracy : {acc:.4f}")
    print(f"  Precision: {prec:.4f}")
    print(f"  Recall   : {rec:.4f}")
    print(f"  F1 Score : {f1:.4f}")
    print("\nClassification Report:")
    print(classification_report(y, y_pred, digits=4))

    # Confusion Matrix
    cm = confusion_matrix(y, y_pred)
    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.title(f"{label} Confusion Matrix")
    plt.show()

evaluate_model_sklearn(best_model, X_train_pad, y_train, label="Train Set")
evaluate_model_sklearn(best_model, X_test_pad, y_test, label="Test Set")
```

**FIND THE BEST THRESHOLD BY F1 SCORE**  
cut off point that turns a model's probability output into a class label  
this is the model output for POSITIVE CLASS  
0.92 -> very confident its positive
0.04 -> very confident its negative
0.48 -> unsure  

the probability must be converted into a discrete class 0 or 1 using a threshold

threshold 0.5 -> the model only predicts "positive" when its more than 50% sure. conservative
threshold 0.37 - the model allows positive even if its just 37% confident. more inclusive. favors recall

```python
import numpy as np
from sklearn.metrics import f1_score

# Initialize tracking
best_thresh = 0.0
best_f1 = 0.0
f1_scores = []

# Evaluate F1 for thresholds between 0.1 and 0.9
thresholds = np.arange(0.1, 0.91, 0.01)
for thresh in thresholds:
    y_pred_thresh = (y_pred_prob > thresh).astype("int32")
    f1 = f1_score(y_test, y_pred_thresh)
    f1_scores.append(f1)
    
    if f1 > best_f1:
        best_f1 = f1
        best_thresh = thresh

print(f"Best threshold: {best_thresh:.2f} with F1 score: {best_f1:.4f}")
```

```python
y_pred_final = (y_pred_prob > 0.52).astype("int32")
#gives maximum f1 performance

from sklearn.metrics import classification_report

print("Final model evaluation at best threshold (0.46):")
print(classification_report(y_test, y_pred_final))
```

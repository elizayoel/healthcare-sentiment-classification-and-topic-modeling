**ARIZONA REVIEW DATASET**

```python
review_path = "C:/Users/eliza/Documents/Semester V/Sentiment Analysis/Arizona - Review/review-Arizona.json"
#read JSON file as line-separated (NDJSON)
review_df = pd.read_json(review_path, lines = True)

review_df.head(10)
```

```python
# Aggregate business metadata by gmap_id before merging
business_agg = business_df.groupby("gmap_id", as_index=False).agg({
    "name": "first",  # Keep first occurrence (assumes same name per gmap_id)
    "address": "first",
    "description": "first",
    "latitude": "first",
    "longitude": "first",
    "category": lambda x: list(set(x.sum())) if x.apply(lambda v: isinstance(v, list)).any() else list(set(x)),
    "avg_rating": "mean",  # Use mean rating if multiple exist
    "num_of_reviews": "sum",
    "price": "first",
    "hours": "first",
    "MISC": "first",
    "state": "first",
    "relative_results": "first",
    "url": "first"
})

# Check if aggregation worked
print(business_agg.head())
```

```python
# Merge with aggregated business data
merged_df = review_df.merge(business_agg, on="gmap_id", how="left")

# convert merged dataset into csv
merged_df.to_csv("arizona_merged_data.csv", index=False)
```

**SEPARATE CATEGORIES IN DATASET INTO DIFFERENT COLUMNS**


```python
import ast

merged_arizona["category"] = merged_arizona["category"].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Expand the list into separate columns
category_expanded = merged_arizona["category"].apply(pd.Series)

# Rename columns as 'category_1', 'category_2', etc.
category_expanded.columns = [f"category_{i+1}" for i in range(category_expanded.shape[1])]

# Concatenate with original dataset
merged_arizona = pd.concat([merged_arizona, category_expanded], axis=1)

# Drop original 'category' column
#cleaned_file.drop(columns=["category"], inplace=True)

merged_arizona
```

**GENERAL HEALTHCARE LABELLING**

```python


# Updated list of healthcare-related business categories
healthcare_categories = [
    "dentist", "cosmetic dentist", "orthodontist", "medical center", "wellness center",
    "pharmacy", "chiropractor", "hearing aid store", "audiologist", "pain management physician",
    "urologist", "doctor", "urology clinic", "medical clinic", "home health care service",
    "herbal medicine store", "vitamin & supplements store", "drug store", "physical fitness program", 
    "dental clinic", "dental implants provider", "denture care center",
    "emergency dental service", "endodontist", "medical group", "pediatric",
    "oral surgeon", "orthopedic surgeon", "orthopedic clinic", "internist", "otolaryngologist",
    "neurologist", "eye care center", "ophthalmologist", "optician", "optometrist", "surgical center",
    "pediatric dentist", "physical therapy clinic", "psychiatrist"
]

# Function to check if a business belongs to the healthcare category
def classify_healthcare(row):
    for i in range(1, 16):  # Assuming category_1 to category_15 exist
        category_col = f"category_{i}"
        if category_col in row and isinstance(row[category_col], str):  # Check if column exists and is a string
            for keyword in healthcare_categories:
                if re.search(rf"\b{keyword}\b", row[category_col], re.IGNORECASE):  # Match whole word
                    return "Healthcare"  # Assign healthcare label
    return "Other"  # Default category if no match

# Apply the function to create a new 'general_labelling' column
merged_arizona["general_labelling"] = merged_arizona.apply(classify_healthcare, axis=1)



```

```python
healthcare_count = merged_arizona["general_labelling"].value_counts().get("Healthcare", 0)
healthcare_count
```

```python
# Filter records where 'general_labelling' is "Healthcare"
healthcare_dataset = merged_arizona[merged_arizona["general_labelling"] == "Healthcare"]

# Define the file path to save the Healthcare dataset
arizona_healthcare_dataset_path = "C:/Users/eliza/Documents/Semester V/Sentiment Analysis/arizona_healthcare_dataset.csv"

# Save the filtered dataset to CSV
healthcare_dataset.to_csv(arizona_healthcare_dataset_path, index=False)
```

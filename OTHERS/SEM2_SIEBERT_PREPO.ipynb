**EMOJI HANDLING**

```python
import emoji
import re

def normalize_emojis(text):
    # demojize emojis
    text = emoji.demojize(text, language='en')

    # add space before/after every emoji token
    text = re.sub(r'(:[^:\s]+?:)', r' \1 ', text)  # add space around :emoji:

    # normalize multiple spaces
    text = re.sub(r'\s+', ' ', text).strip()

    return text
```

```python
# example
emoji_df = 'Amazing service! 😍🔥 :('
emoji_df = normalize_emojis(emoji_df)
emoji_df
```

```python
test_emoji_sentences = [
    "Great service!😍🔥Awesome!",                       # No space between emojis or words
    "I love it 😍👍",                                   # Proper spacing
    "Horrible experience.😡😠Will not return!",         # No space after sentence + emojis
    "Okay. 😐",                                          # Emoji after period
    "This is bad. 😢",                                   # Space before emoji
    "So cute!😊    🥰",                                  # Multiple spaces
    ":)",                                                # ASCII emoji
    "Well done👏👏👏",                                   # Emoji clapped together
    "Clean place.👍Fast service.👌",                    # Emojis between sentences, no punctuation
    "Just wow :smiling_face:🔥",                        # Already demojized
    "Friendly staff😊. Clean room😷. Quiet at night😴",  # Emojis embedded
    "😍  😍😍",                                           # Emojis with various spacing
    "Worst visit ever.😡  Not coming back😤.",           # Emoji + period spacing test
]

for sentence in test_emoji_sentences:
    cleaned = normalize_emojis(sentence)
    print(f"Original: {sentence}")
    print(f"Cleaned:  {cleaned}\n")
```

**EXPAND CONTRACTIONS**

```python
import contractions

contractions.fix("I can't believe it's already done. She's ready, isn't she?")
```

```python
import contractions
def cont_expand(text):
    if isinstance(text, str) and text.strip():
        try:
            return contractions.fix(text)
        except IndexError:
            return text  # fallback to original text if expansion fails
    else:
        return ""
    
df['expanded_text'] = df['emoji_text'].apply(cont_expand)
df
```

**SENTENCE TOKENIZATION ON EXPANDED TEXT**

```python
import nltk
from nltk.tokenize import sent_tokenize
df['sentences'] = df['expanded_text'].apply(sent_tokenize)
df
```

**WORD TOKENIZATION ON SENTENCES**

```python
from nltk.tokenize import word_tokenize
nltk.download('punkt')

def tokenize_words(sentences):
    return [word_tokenize(sentence) for sentence in sentences]

df['word_tokens'] = df['sentences'].apply(tokenize_words)
```

**LOWERCASING**

```python
df['word_tokens'] = df['word_tokens'].apply(
    lambda sentence_list: [[word.lower() for word in sentence] for sentence in sentence_list]
)
df
```

**REMOVE PUNCTUATION**

```python
import string

def remove_punctuation(token_lists):
    return [[word for word in sentence if word not in string.punctuation] for sentence in token_lists]

df['punctuation_removed'] = df['word_tokens'].apply(remove_punctuation)
df
```

**POS TAGGING USING SPACY** 

```python
from nltk import pos_tag

nltk.download('averaged_perceptron_tagger')

def tag_pos(sentence_list):
    return [pos_tag(sentence) for sentence in sentence_list]

df['pos_tags'] = df['punctuation_removed'].apply(tag_pos)
```

**LEMMATIZATION USING POS TAGGING**

```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

nltk.download('wordnet')
nltk.download('omw-1.4')

# convert Treebank tag to WordNet tag
def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN #default to noun if unknown
    

# initialize lemmatizer
lemmatizer = WordNetLemmatizer()

def lemmatize_with_pos(pos_tagged_sentences):
    return [
        [lemmatizer.lemmatize(word, get_wordnet_pos(pos_tag)) for word, pos_tag in sentence]
        for sentence in pos_tagged_sentences
    ]

df['lemmatized'] = df['pos_tags'].apply(lemmatize_with_pos)
```

**REMOVE STOPWORDS**

```python
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')

stop_words = set(stopwords.words('english'))
# add custom stop words = negation
negation_words = {"not", "no", "nor"}
# negation wont be removed
stop_words = stop_words - negation_words

def remove_stopwords(tokenized_sentences):
    return [
        [word for word in sentence if word not in stop_words]
        for sentence in tokenized_sentences
    ]

df['no_stopwords'] = df['lemmatized'].apply(remove_stopwords)
df
```

**REMOVE REDUNDANT PUNCTUATION**

```python
import re

def remove_redundant_punct(sentences):
    cleaned = []
    for sentence in sentences:

        filtered = [token for token in sentence if not re.fullmatch(r'[^\w\s]+', token)]
        cleaned.append(filtered)
    return cleaned

df['lemma_fix'] = df['lemmatized'].apply(remove_redundant_punct)
```

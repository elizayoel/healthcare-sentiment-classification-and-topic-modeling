```python
import pandas as pd
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline
# to monitor progress
from tqdm import tqdm

tqdm.pandas()

# dataset has been loaded

# load pretrained model & tokenizer
model_name = 'cardiffnlp/twitter-roberta-base-sentiment'
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# create sentiment pipeline
sentiment_pipe = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)

# apply to text column
def get_sentiment(text):
    if pd.isna(text) or not isinstance(text, str) or text.strip() == '':
        return {'label': 'unknown', 'score': 0.0}
    try:
        result = sentiment_pipe(text[:512])[0]  # Truncate long texts to 512 tokens
        return result
    except Exception as e:
        return {'label': 'error', 'score': 0.0}

# Apply BERT to the cleaned text
text_df['sentiment_result'] = text_df['cleaned_translated_text'].progress_apply(get_sentiment)
text_df['sentiment_label'] = text_df['sentiment_result'].apply(lambda x: x['label'])
text_df['sentiment_score'] = text_df['sentiment_result'].apply(lambda x: x['score'])
text_df.drop(columns=['sentiment_result'], inplace=True)

```

```python
import contractions
def safe_expand(text):
    if isinstance(text, str) and text.strip():
        try:
            return contractions.fix(text)
        except IndexError:
            return text  # fallback to original text if expansion fails
    else:
        return ""
    
bert_df['expanded_text'] = bert_df['prepo_text'].apply(safe_expand)
```

```python
bert_df
```

**sentence tokenization**

```python
import nltk
from nltk.tokenize import sent_tokenize
bert_df['sentences'] = bert_df['expanded_text'].apply(sent_tokenize)
bert_df
```

**word tokenization**

```python
from nltk.tokenize import word_tokenize
nltk.download('punkt')
# Tokenize sentences into words
def tokenize_words(sentences):
    return [word_tokenize(sentence) for sentence in sentences]

bert_df['word_tokens'] = bert_df['sentences'].apply(tokenize_words)
```

**lowercasing**

```python
bert_df['word_tokens'] = bert_df['word_tokens'].apply(
    lambda sentence_list: [[word.lower() for word in sentence] for sentence in sentence_list]
)
```

```python
bert_df
```

**remove punctuation**

```python
import string

def remove_punctuation(token_lists):
    return [[word for word in sentence if word not in string.punctuation] for sentence in token_lists]

bert_df['punctuation_removed'] = bert_df['word_tokens'].apply(remove_punctuation)
```

```python
bert_df
```

**pos tagging using nltk**

```python
from nltk import pos_tag

nltk.download('averaged_perceptron_tagger')

def tag_pos(sentence_list):
    return [pos_tag(sentence) for sentence in sentence_list]

bert_df['pos_tags'] = bert_df['punctuation_removed'].apply(tag_pos)
```

**lemmatization**

```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

nltk.download('wordnet')
nltk.download('omw-1.4')

# convert Treebank tag to WordNet tag
def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN #default to noun if unknown
    

# initialize lemmatizer
lemmatizer = WordNetLemmatizer()

def lemmatize_with_pos(pos_tagged_sentences):
    return [
        [lemmatizer.lemmatize(word, get_wordnet_pos(pos_tag)) for word, pos_tag in sentence]
        for sentence in pos_tagged_sentences
    ]

bert_df['lemmatized_with_pos'] = bert_df['pos_tags'].apply(lemmatize_with_pos)
```

```python
bert_df
```

```python
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')

stop_words = set(stopwords.words('english'))
# add custom stop words = negation
negation_words = {"not", "no", "nor"}
stop_words = stop_words - negation_words

def remove_stopwords(tokenized_sentences):
    return [
        [word for word in sentence if word not in stop_words]
        for sentence in tokenized_sentences
    ]

bert_df['no_stopwords'] = bert_df['lemmatized_with_pos'].apply(remove_stopwords)
```

**rejoin tokens - no stopwords**

```python
bert_df['final_no_stopword'] = bert_df['no_stopwords'].apply(
    lambda sentence_list: ' '.join([word for sentence in sentence_list for word in sentence])
)
```

```python
bert_df
```

**rejoin tokens with stopwords**

```python
bert_df['final_with_stopword'] = bert_df['lemmatized_with_pos'].apply(
    lambda sentence_list: ' '.join([word for sentence in sentence_list for word in sentence])
)
```

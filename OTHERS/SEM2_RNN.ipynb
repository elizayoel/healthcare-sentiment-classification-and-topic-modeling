```python
from tensorflow.keras.preprocessing.text import Tokenizer

# Set vocab size
vocab_size = 10000
tokenizer = Tokenizer(num_words=vocab_size, oov_token="<OOV>")

# Fit on training data
tokenizer.fit_on_texts(X_train_resampled)

# Convert text to sequences
X_train_seq = tokenizer.texts_to_sequences(X_train_resampled)
X_test_seq = tokenizer.texts_to_sequences(X_test.tolist())  # use the original test set
```

**PADDING**

**CHECKING REVIEW LENGTH TO DETERMINE MAX_LEN**

```python
review_lengths = [len(seq) for seq in X_train_seq]
import matplotlib.pyplot as plt
plt.hist(review_lengths, bins=50)
plt.xlabel("Length of review (tokens)")
plt.ylabel("Number of reviews")
plt.show()
```

```python
import numpy as np

review_lengths = [len(seq) for seq in X_train_seq]
np.percentile(review_lengths, [50, 75, 90, 95])
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Define max length based on your percentile analysis
max_len = 100

# Pad training and test sequences
X_train_pad = pad_sequences(X_train_seq, maxlen=max_len, padding='post', truncating='post')
X_test_pad = pad_sequences(X_test_seq, maxlen=max_len, padding='post', truncating='post')
```

**GLOVE EMBEDDING**

```python
embedding_index = {}
with open("glove.6B.100d.txt", encoding='utf8') as f:
    for line in f:
        values = line.split()
        word = values[0]
        vector = np.asarray(values[1:], dtype='float32')
        embedding_index[word] = vector
```

```python
# create embedding matrix
embedding_dim = 100
word_index = tokenizer.word_index
embedding_matrix = np.zeros((vocab_size, embedding_dim))

for word, i in word_index.items():
    if i < vocab_size:
        embedding_vector = embedding_index.get(word)
        if embedding_vector is not None:
            embedding_matrix[i] = embedding_vector
```

```python
from tensorflow.keras.layers import Embedding

# create embedding layer
embedding_layer = Embedding(
    input_dim=vocab_size,
    output_dim=embedding_dim,
    weights=[embedding_matrix],
    trainable=False  # Set to True if you want to fine-tune GloVe
)
```

**BUILD LSTM**

```python
from tensorflow.keras.layers import LSTM, Dropout

model = Sequential()
model.add(embedding_layer)
model.add(LSTM(64, return_sequences=False))
model.add(Dropout(0.3))  # helps regularize
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)
```

```python
model.summary()
```

```python
history = model.fit(
    X_train_pad,
    y_train_resampled,
    epochs=5,
    batch_size=32,
    validation_data=(X_test_pad, y_test)
)
```

**PREDICT LABELS ON TEST SET**

```python
import numpy as np

# predict probabilities
y_pred_probs = model.predict(X_test_pad)

# convert probabilities to binary class predictions (0 or 1)
y_pred = (y_pred_probs > 0.5).astype("int32")
```

```python
from sklearn.metrics import confusion_matrix, classification_report, f1_score
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['NEGATIVE', 'POSITIVE'], yticklabels=['NEGATIVE', 'POSITIVE'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
print(classification_report(y_test, y_pred, target_names=['NEGATIVE', 'POSITIVE']))
```

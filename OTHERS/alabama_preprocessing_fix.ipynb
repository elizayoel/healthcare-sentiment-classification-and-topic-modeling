```python
#explode the 'category' column to separate multiple categories into individual rows
business_df_exploded = business_df.explode('category')

#get unique categories
unique_categories = business_df_exploded['category'].unique()

print(unique_categories)
```

```python
#user review has been loaded = dfr
#business metadata has been loaded = dfm
#merge both df based on gmap_id


merged_df = review_df.merge(business_df_exploded, on = "gmap_id", how = "inner")
```

```python
merged_df.head()
```

```python
# Aggregate business metadata by gmap_id before merging
business_agg = business_df.groupby("gmap_id", as_index=False).agg({
    "name": "first",  # Keep first occurrence (assumes same name per gmap_id)
    "address": "first",
    "description": "first",
    "latitude": "first",
    "longitude": "first",
    "category": lambda x: list(set(x.sum())) if x.apply(lambda v: isinstance(v, list)).any() else list(set(x)),
    "avg_rating": "mean",  # Use mean rating if multiple exist
    "num_of_reviews": "sum",
    "price": "first",
    "hours": "first",
    "MISC": "first",
    "state": "first",
    "relative_results": "first",
    "url": "first"
})

# Check if aggregation worked
print(business_agg.head())
```

```python
# Merge with aggregated business data
merged_df = review_df.merge(business_agg, on="gmap_id", how="left")

# convert merged dataset into csv
merged_df.to_csv("merged_data.csv", index=False)
```

**SEPARATE CATEGORIES IN DATASET INTO DIFFERENT COLUMNS**

```python
#run cleaned_file

#convert 'category' column from string to list (if needed)
import ast
cleaned_file["category"] = cleaned_file["category"].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Expand the list into separate columns
category_expanded = cleaned_file["category"].apply(pd.Series)

# Rename columns as 'category_1', 'category_2', etc.
category_expanded.columns = [f"category_{i+1}" for i in range(category_expanded.shape[1])]

# Concatenate with original dataset
cleaned_file = pd.concat([cleaned_file, category_expanded], axis=1)

# Drop original 'category' column
#cleaned_file.drop(columns=["category"], inplace=True)

cleaned_file
```

```python
import re
```

```python
# List of healthcare-related business categories
healthcare_categories = [
    "rehabilitation center", "nursing home", "doctor", "medical clinic", "retirement home", 
    "addiction treatment center", "dental implants periodontist", "medical center",
    "contact lenses supplier", "eye care center", "pediatric dentist", "dentist", "internist", 
    "podiatrist", "blood testing center", "diagnostic center", "paternity testing service",
    "STD testing service", "drug testing service", "medical laboratory", "massage therapist", 
    "sports massage therapist", "neurologist", "medical school", "dental school", "skin care clinic",
    "obstetrician-gynecologist"
]

# Function to check if a business belongs to the healthcare category
def classify_healthcare(row):
    for i in range(1, 16):  # Assuming category_1 to category_15 exist
        category_col = f"category_{i}"
        if category_col in row and isinstance(row[category_col], str):  # Check if column exists and is a string
            for keyword in healthcare_categories:
                if re.search(rf"\b{keyword}\b", row[category_col], re.IGNORECASE):  # Match whole word
                    return "Healthcare"  # Assign healthcare label
    return "Other"  # Default category if no match

# Apply the function to create a new 'general_labelling' column
cleaned_file["general_labelling"] = cleaned_file.apply(classify_healthcare, axis=1)
cleaned_file
```

```python
healthcare_count = cleaned_file["general_labelling"].value_counts().get("Healthcare", 0)
healthcare_count
```

```python
# Filter records where 'general_labelling' is "Healthcare"
healthcare_dataset = cleaned_file[cleaned_file["general_labelling"] == "Healthcare"]

# Define the file path to save the Healthcare dataset
alabama_healthcare_dataset_path = "C:/Users/eliza/Documents/Semester V/Sentiment Analysis/alabama_healthcare_dataset.csv"

# Save the filtered dataset to CSV
healthcare_dataset.to_csv(alabama_healthcare_dataset_path, index=False)
```

```python
import pandas as pd

# List of all dataframe variable names
states = [
    "alabama_arkansas", "kentucky", "louisiana", "nevada", "maryland",
    "massachusetts", "minnesota", "utah", "southcarolina", "oregon", "wisconsin"
]

# Collect all df into a list
dfs = []
for state in states:
    df = globals().get(state)
    if df is not None:
        df['state'] = state 
        dfs.append(df)
    else:
        print(f"all_merge.ipynbWarning: {state} not found.")


combined_df = pd.concat(dfs, ignore_index=True, sort=False)
```

**reduce rating 5 size**

```python
# Filter rating 5 reviews
rating5_df = combined_df[combined_df['rating'] == 5]

# Randomly sample 500,000 of them
rating5_sampled = rating5_df.sample(n=500000, random_state=42)

# Filter reviews that are NOT rating 5
other_ratings_df = combined_df[combined_df['rating'] != 5]

combined_reduced_df = pd.concat([rating5_sampled, other_ratings_df], ignore_index=True)

combined_reduced_df = combined_reduced_df.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
import re

def remove_translated_text(text):
    if isinstance(text,str):
        # Remove '(Translated by Google)' if present
        cleaned = re.sub(r'\(Translated by Google\)\s*', '', text)
        # Remove '(Original)' and everything after it
        cleaned = re.split(r'\(Original\)', cleaned)[0]
        return cleaned.strip()  # Remove leading and trailing whitespaces
    return text  # Return original if it's not a string


#apply function and create new column
prepo_df['translated_text'] = prepo_df['original_text'].apply(remove_translated_text)
```

```python
# Filter records where the text starts with '(Translated by Google)'
translated_records = prepo_df[prepo_df['translated_text'].str.contains(r'^\(Translated by Google\)', regex=True, na=False)]

print(f"Total records starting with '(Translated by Google)': {translated_records.shape[0]}")
```

```python
import re

def remove_translated_text(text):
    if isinstance(text,str):
        # Remove '(Translated by Google)' if present
        cleaned = re.sub(r'\(Translated by Google\)\s*', '', text)
        # Remove '(Original)' and everything after it
        cleaned = re.split(r'\(Original\)', cleaned)[0]
        return cleaned.strip()  # Remove leading and trailing whitespaces
    return text  # Return original if it's not a string


#apply function and create new column
df['translated_text'] = df['original_text'].apply(remove_translated_text)
```

**REDUCE OVER DUPLICATED COUNT**

```python
# Count text frequencies
text_counts = newdf['translated_text'].value_counts()

# Get duplicated and unique texts
duplicated_texts = text_counts[text_counts > 1].index
unique_texts = text_counts[text_counts == 1].index

# Filter the full DataFrame
duplicates_df = newdf[newdf['translated_text'].isin(duplicated_texts)]
unique_df = newdf[newdf['translated_text'].isin(unique_texts)]
```

```python
reduced_duplicates_df = (
    duplicates_df.groupby('translated_text')
    .apply(lambda x: x.sample(n=min(len(x), 5), random_state=42))
    .reset_index(drop=True)
)
```

```python
mindup_df = pd.concat([unique_df, reduced_duplicates_df], ignore_index=True)
```

```python
#recheck
duplicate_counts = mindup_df['translated_text'].value_counts()
duplicates_only = duplicate_counts[duplicate_counts > 1]
duplicates_only
```

## **HANDLE REVIEW WITH SMALL WORD COUNT**

```python
mindup_df['wcount'] = mindup_df['translated_text'].str.split().str.len()
mindup_df
```

```python
filter_df = mindup_df[mindup_df['wcount'] == 2]
filter_df
```

```python
import re

def is_low_value(text):
    if pd.isna(text) or not isinstance(text, str):
        return True

    stripped = text.strip()
    
    # Remove if word count is 1 or 2
    word_count = len(stripped.split())
    if word_count <= 2:
        return True

    # Remove if only emojis or symbols (no letters or digits)
    if not re.search(r'[a-zA-Z0-9]', stripped):
        return True

    return False
```

```python
#filter out low value rows
filtered_df = mindup_df[~mindup_df['translated_text'].apply(is_low_value)].copy()
```

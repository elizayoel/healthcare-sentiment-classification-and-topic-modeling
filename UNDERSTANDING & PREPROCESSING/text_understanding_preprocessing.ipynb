```python
import re

# Function to clean the text and remove everything after '(Original)'
def clean_translated_text(text):
    if isinstance(text, str):
        # Remove '(Translated by Google)' if present
        cleaned = re.sub(r'\(Translated by Google\)\s*', '', text)
        # Remove '(Original)' and everything after it
        cleaned = re.split(r'\(Original\)', cleaned)[0]
        return cleaned.strip()  # Remove leading and trailing whitespaces
    return text  # Return original if it's not a string

# Apply the function and create a new column
text_df['cleaned_translated_text'] = text_df['text'].apply(clean_translated_text)


```

## **NLTK**

**SENTENCE SEGMENTATION**

```python
from nltk.tokenize import sent_tokenize

# Function to apply sentence tokenization
def tokenize_sentences(text):
    if isinstance(text, str):
        return sent_tokenize(text)
    return []

# Apply sentence tokenization and store it in a new column
text_df['nltk_sentence_tokenized'] = text_df['cleaned_translated_text'].apply(tokenize_sentences)

```

```python
text_df
```

**WORD TOKENIZATION**

```python
from nltk.tokenize import word_tokenize

# Function to perform word segmentation on each sentence
def word_segmentation(sentences):
    if isinstance(sentences, list):
        return [word_tokenize(sentence) for sentence in sentences]
    return []

# Apply word segmentation and store in a new column
text_df['nltk_word_tokenized_from_sentences'] = text_df['nltk_sentence_tokenized'].apply(word_segmentation)

```

```python
text_df
```

**LOWERCASING**

```python
# Function to apply lowercasing to word tokenized sentences
def lowercase_tokens(tokenized_sentences):
    if isinstance(tokenized_sentences, list):
        return [[word.lower() for word in sentence] for sentence in tokenized_sentences]
    return []

# Apply lowercasing and store in a new column
text_df['nltk_lowercased_tokens'] = text_df['nltk_word_tokenized_from_sentences'].apply(lowercase_tokens)
```

```python
text_df
```

**REMOVING PUNCTUATION**

```python
import string

# Define punctuation characters
punctuations = set(string.punctuation)

# Function to remove punctuation from tokenized sentences
def remove_punctuation(tokenized_sentences):
    if isinstance(tokenized_sentences, list):
        return [[word for word in sentence if word not in punctuations] for sentence in tokenized_sentences]
    return []

# Apply punctuation removal and store it in a new column
text_df['nltk_no_punctuation_tokens'] = text_df['nltk_lowercased_tokens'].apply(remove_punctuation)

```

```python
text_df
```

**POS TAGGING**  


```python
import nltk

# Function to perform POS tagging on tokenized sentences
def pos_tagging(tokenized_sentences):
    if isinstance(tokenized_sentences, list):
        return [nltk.pos_tag(sentence) for sentence in tokenized_sentences]
    return []

# Apply POS tagging and store in a new column
text_df['nltk_pos_tagged_tokens'] = text_df['nltk_no_punctuation_tokens'].apply(pos_tagging)

```

```python
text_df
```

**LEMMATIZATION AFTER POS TAGGING**

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

# Initialize the lemmatizer
lemmatizer = WordNetLemmatizer()

# Function to map NLTK POS tags to WordNet POS tags
def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN  # Default to noun if not recognized

# Function to perform lemmatization using POS tagging
def lemmatize_with_pos(pos_tagged_sentences):
    if isinstance(pos_tagged_sentences, list):
        return [
            [lemmatizer.lemmatize(word, get_wordnet_pos(pos)) for word, pos in sentence]
            for sentence in pos_tagged_sentences
        ]
    return []

# Apply the lemmatization function and store the result in a new column
text_df['nltk_lemmatized_tokens'] = text_df['nltk_pos_tagged_tokens'].apply(lemmatize_with_pos)

```

```python
text_df
```

```python
import nltk
from nltk.corpus import stopwords

# Download the stopwords dataset
nltk.download('stopwords')

# Define English stopwords
stop_words = set(stopwords.words('english'))

# Function to remove stopwords from lemmatized tokens
def remove_stopwords(tokenized_sentences):
    if isinstance(tokenized_sentences, list):
        return [
            [word for word in sentence if word not in stop_words]
            for sentence in tokenized_sentences
        ]
    return []

# Apply stopword removal and store the result in a new column
text_df['nltk_no_stopwords_tokens'] = text_df['nltk_lemmatized_tokens'].apply(remove_stopwords)
```

```python
text_df
```

**REJOIN TOKENS AS STRINGS**

```python
# Function to rejoin tokens into strings
def rejoin_tokens(tokenized_sentences):
    if isinstance(tokenized_sentences, list):
        return [' '.join(sentence) for sentence in tokenized_sentences]
    return ''

# Apply the function and store the results in a new column
text_df['nltk_final_cleaned_text'] = text_df['nltk_no_stopwords_tokens'].apply(lambda sentences: ' '.join([' '.join(sentence) for sentence in sentences]))

```

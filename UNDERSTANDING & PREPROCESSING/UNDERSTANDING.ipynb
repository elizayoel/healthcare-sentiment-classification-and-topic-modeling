```python

incorrect_state_label = 'alabama_arkansas'

alabama_gmap_ids = set(alabama['gmap_id'])
arkansas_gmap_ids = set(arkansas['gmap_id'])
alabama_fix_mask = (df['state'] == incorrect_state_label) & \
                   (df['gmap_id'].isin(alabama_gmap_ids))

df.loc[alabama_fix_mask, 'state'] = 'Alabama'
arkansas_fix_mask = (df['state'] == incorrect_state_label) & \
                    (df['gmap_id'].isin(arkansas_gmap_ids))

df.loc[arkansas_fix_mask, 'state'] = 'Arkansas'


# --- Step 4: Verify the fix ---
print("Verification:")
remaining_incorrect = df[df['state'] == incorrect_state_label].shape[0]
print(f"Number of rows still labeled as '{incorrect_state_label}': {remaining_incorrect}")

print("\nChecking the new counts for Alabama and Arkansas:")
print(df['state'].value_counts().loc[['Alabama', 'Arkansas']])
```

**TEXT LENGTH ANALYSIS**

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import ScalarFormatter  

# Create character and word length columns
df['text_length_char'] = df['text'].astype(str).str.len()
df['text_length_word'] = df['text'].astype(str).str.split().str.len()

# === Character length ===
plt.figure(figsize=(8, 4))
sns.histplot(df['text_length_char'], bins=50, kde=False)
plt.title("Review Length (Characters)")
plt.xlabel("Character Count")
plt.ylabel("Frequency")
plt.xlim(0, 1000)
plt.gca().yaxis.set_major_formatter(ScalarFormatter())
plt.tight_layout()
plt.show()

# === Word length ===
plt.figure(figsize=(8, 4))
sns.histplot(df['text_length_word'], bins=50, kde=False, color='green')
plt.title("Review Length (Words)")
plt.xlabel("Word Count")
plt.ylabel("Frequency")
plt.xlim(0, 200)
plt.gca().yaxis.set_major_formatter(ScalarFormatter())  
plt.tight_layout()
plt.show()

```

## **WORD CLOUD**

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

# Combine all text into a single string
text_data = ' '.join(df['text'].dropna().astype(str).values)

# Define stopwords
custom_stopwords = set(STOPWORDS)
custom_stopwords.update(["I", "The", "It", "They", "We", "This", "That", "Also"])

# Create the word cloud
wordcloud = WordCloud(
    width=1200,
    height=600,
    background_color='white',
    stopwords=custom_stopwords,
    max_words=200,
    collocations=True
).generate(text_data)

# Display it
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Most Frequent Words in Review Texts")
plt.tight_layout()
plt.show()
```

```python
from collections import Counter
import pandas as pd
import re

# Combine all text into a single string
all_text = ' '.join(df['text'].dropna().astype(str).values).lower()
# Remove punctuation and numbers
all_text = re.sub(r'[^a-z\s]', '', all_text)
# Tokenize
words = all_text.split()
# Optional: remove stopwords
from wordcloud import STOPWORDS
stopwords = set(STOPWORDS)
filtered_words = [word for word in words if word not in stopwords and len(word) > 2]
# Count word frequencies
word_counts = Counter(filtered_words)
top_words = word_counts.most_common(20)
# Convert to DataFrame
word_freq_df = pd.DataFrame(top_words, columns=['word', 'count'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.barplot(x='count', y='word', data=word_freq_df, palette='magma')
plt.title("Top 20 Most Frequent Words in Review Text")
plt.xlabel("Frequency")
plt.ylabel("Word")
plt.tight_layout()
plt.show()
```

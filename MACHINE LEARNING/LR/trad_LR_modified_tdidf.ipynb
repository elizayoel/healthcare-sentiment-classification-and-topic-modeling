**EXTRACT FEATURES AND LABELS**

```python
#extract features and labels
X_train_texts = train_df['final_with_stopword'].astype(str).tolist()
y_train = train_df['label'].values
X_test_texts = test_df['final_with_stopword'].astype(str).tolist()
y_test = test_df['label'].values
```

**TEXT VECTORIZATION (TF IDF)**

**parameters**  
1. max_features  
no default  
limit vocabulary to top-N most frequent words  
5k-20k for large datasets 
2. ngram_range  
unigram, bigram, trigram  
3. min_df  
default = 1 = ignore words in fewer than min_df documents  
use 2 or 3 to remove rare noise  
4. max_df  
default = 1.0  
ignore words more than max_df docs  
use 0.9 or 0.95 to remove very common stopwords  
5. sublinear_tf  
default = false

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=10000, 
                        ngram_range=(1,2),
                        min_df = 2,
                        max_df = 0.90,
                        sublinear_tf=True
                        )
X_train_tfidf = tfidf.fit_transform(X_train_texts)
X_test_tfidf = tfidf.transform(X_test_texts)

#save vectorizer 
# import pickle
# with open("tfidf_vectorizer15000_ngram3.pkl", "wb") as f:
#     pickle.dump(tfidf, f)
```

**IMPORT AND TRAIN MODEL**

```python
from sklearn.linear_model import LogisticRegression

# Initialize the model
lr_model = LogisticRegression(max_iter=1000)

# Train the model
lr_model.fit(X_train_tfidf, y_train)
```

**MAKE PREDICTIONS ON TEST SET**

```python
#predict class label
y_pred = lr_model.predict(X_test_tfidf)
#predict class probabilities for log loss
y_proba = lr_model.predict_proba(X_test_tfidf)
```

**EVALUATE THE MODEL**

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, log_loss
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

def evaluate_model_sklearn(model, X, y, label="Test Set"):
    # Predict probabilities and round to binary (0 or 1)
    y_pred_prob = model.predict(X)
    y_pred = (y_pred_prob > 0.5).astype("int32")

    # Metrics
    acc = accuracy_score(y, y_pred)
    prec = precision_score(y, y_pred)
    rec = recall_score(y, y_pred)
    f1 = f1_score(y, y_pred)
    logloss = log_loss(y, y_pred_prob)

    print(f" {label} Evaluation")
    print(f"  Accuracy : {acc:.4f}")
    print(f"  Precision: {prec:.4f}")
    print(f"  Recall   : {rec:.4f}")
    print(f"  F1 Score : {f1:.4f}")
    print(f"  Log Loss : {logloss:.4f}")
    print("\nClassification Report:")
    print(classification_report(y, y_pred, digits=4))

    # Confusion Matrix
    cm = confusion_matrix(y, y_pred)
    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.title(f"{label} Confusion Matrix")
    plt.show()

evaluate_model_sklearn(lr_model, X_train_tfidf, y_train, label="Train Set")
evaluate_model_sklearn(lr_model, X_test_tfidf, y_test, label="Test Set")

```

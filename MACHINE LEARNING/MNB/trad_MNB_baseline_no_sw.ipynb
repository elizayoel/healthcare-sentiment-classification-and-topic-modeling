**LOAD DATASET**

```python
import pandas as pd
train_df = pd.read_csv("data/train_res.csv")
test_df = pd.read_csv("data/test.csv")
```

**EXTRACT FEATURES AND LABELS**

```python
X_train_texts = train_df['final_no_stopword'].astype(str).tolist()
y_train = train_df['label'].values

X_test_texts = test_df['final_no_stopword'].astype(str).tolist()
y_test = test_df['label'].values
```

**LOAD SAVED TF-IDF VECTORIZER**  


```python
import pickle
with open("tfidf_vectorizer.pkl", "rb") as f:
    tfidf = pickle.load(f)
```

**TRANSFORM TEXT**

```python
X_train_tfidf = tfidf.transform(X_train_texts)
X_test_tfidf = tfidf.transform(X_test_texts)
```

**LOAD MODEL**

```python
from sklearn.naive_bayes import MultinomialNB
nb_model = MultinomialNB()
```

**FIT THE MODEL**

```python
import time
start = time.time()
nb_model.fit(X_train_tfidf, y_train)
print(f"Time taken: {(time.time() - start)/60:.2f} minutes")
```

**SAVE BEST MODEL**

```python
import pickle
with open("model_MNB_baseline_noSW.pkl", "wb") as f:
    pickle.dump(nb_model, f)
```

**PREDICT ON TEST SET**

```python
# Predict on test set
y_pred = nb_model.predict(X_test_tfidf)
y_proba = nb_model.predict_proba(X_test_tfidf)
```

**EVALUATE THE MODEL**

```python
import pandas as pd
import pickle
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    classification_report, confusion_matrix
)

# Evaluation function
def evaluate_model_sklearn(y_true, y_pred, y_proba, label="Set"):
    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred)
    rec = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)

    print(f"--- {label} Evaluation ---")
    print(f"Accuracy : {acc:.4f}")
    print(f"Precision: {prec:.4f}")
    print(f"Recall   : {rec:.4f}")
    print(f"F1 Score : {f1:.4f}")
    print("Classification Report:")
    print(classification_report(y_true, y_pred, digits=4))

    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.title(f"{label} Confusion Matrix")
    plt.tight_layout()
    plt.show()


```

```python
#train set evaluation
y_train_pred = nb_model.predict(X_train_tfidf)
y_train_proba = nb_model.predict_proba(X_train_tfidf)
# Evaluation
evaluate_model_sklearn(y_train, y_train_pred, y_train_proba, label="Train Set")

# same for test
y_test_pred = nb_model.predict(X_test_tfidf)
y_test_proba = nb_model.predict_proba(X_test_tfidf)
evaluate_model_sklearn(y_test, y_test_pred, y_test_proba, label="Test Set")
```

```python
from sklearn.metrics import log_loss

# Step 1: Predict class probabilities on train set (using TF-IDF input)
# This returns shape (n_samples, 2): [:, 0] = class 0, [:, 1] = class 1
y_train_proba = nb_model.predict_proba(X_train_tfidf)

# Step 2: Use only the probabilities for the positive class (class 1)
y_train_prob_class1 = y_train_proba[:, 1]

# Step 3: Compute binary log loss
logloss = log_loss(y_train, y_train_prob_class1)

# Step 4: Print the result
print(f"Training Log Loss (Binary, class 1): {logloss:.4f}")
```

```python
from sklearn.metrics import log_loss

# Step 1: Predict class probabilities on test set (using TF-IDF input)
# This returns shape (n_samples, 2): [:, 0] = class 0, [:, 1] = class 1
y_test_proba = nb_model.predict_proba(X_test_tfidf)

# Step 2: Use only the probabilities for the positive class (class 1)
y_test_prob_class1 = y_test_proba[:, 1]

# Step 3: Compute binary log loss
logloss = log_loss(y_test, y_test_prob_class1)

# Step 4: Print the result
print(f"Test Log Loss (Binary, class 1): {logloss:.4f}")
```

```python
from sklearn.metrics import roc_curve, auc, roc_auc_score
import matplotlib.pyplot as plt

# Step 1: Get predicted probabilities for the positive class (class 1)
y_test_proba = nb_model.predict_proba(X_test_tfidf)[:, 1]

# Step 2: Compute ROC curve and ROC AUC score
fpr, tpr, thresholds = roc_curve(y_test, y_test_proba)
roc_auc = roc_auc_score(y_test, y_test_proba)

# Step 3: Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f"ROC Curve (AUC = {roc_auc:.4f})", color='darkorange', lw=2)
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')  # random guess line
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate (Recall)")
plt.title("ROC Curve - Multinomial Naive Bayes")
plt.legend(loc="lower right")
plt.grid(True)
plt.tight_layout()
plt.show()
```
